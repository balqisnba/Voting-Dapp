```python
import sqlite3
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
conn = sqlite3.connect(':memory:')
```

```python
conn.execute('''
CREATE TABLE transactions (
    customer_id INTEGER,
    total_spent REAL,
    transaction_date TEXT
)
''')
```

```python
transaction_date = [
    (1, 250.0, '2023-01-05'),
    (2, 400.0, '2023-02-10'),
    (3, 150.0, '2023-03-15'),
    (4, None, '2023-04-20'),
    (2, 400.0, '2023-02-10')
]
```

```python
conn.executemany('INSERT INTO transactions VALUES (?, ?, ?)', transaction_date)
conn.commit()
```

```python
df_csv = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 5],
    'age': [25, 30, np.nan, 45,50],
    'bmi' : [22.5, 27.8, 30.1, 28.0, 26.5],
    'activity_level': ['High', 'Medium', 'Low', 'Medium', 'High']
})
```

```python
df_csv.to_csv('customer_lifestyle.csv', index=False)
df_csv = pd.read_csv('customer_lifestyle.csv')
df_csv.head()
```

```python
query = "SELECT * FROM transactions"
df = pd.read_sql_query(query, conn)
df.head()
```

```python
print("SQL missing values:\n", df.isnull().sum())
print("CSV missing values:\n", df_csv.isnull().sum())
```

```python
df['total_spent'].fillna(df['total_spent'].mean(), inplace=True)
df_csv['age'].fillna(df_csv['age'].mean(), inplace=True)
```

```python
df.drop_duplicates(inplace=True)
df_csv.drop_duplicates(inplace=True)
```

```python
df_csv = df_csv = df_csv[(df_csv['age'] >=0) & (df_csv['age'] <= 100)]
```

```python
df_csv['activity_level'] = df_csv['activity_level'].str.lower().str.strip()
```

```python
scaler = StandardScaler()
df_csv[['age', 'bmi']] = scaler.fit_transform(df_csv[['age', 'bmi']])
```

```python
df_csv = pd.get_dummies(df_csv, columns=['activity_level'], drop_first=True)
```

```python
df_final = pd.merge(df, df_csv, on='customer_id', how='inner')
df_final.head()
```

```python
print("Missing values after merge:\n", df_final.isnull().sum())
print("Duplicate rows after merge:\n", df_final.duplicated().sum())
```

```python
df_final.describe()
```

```python
df_final.to_csv('final_data.csv', index=False)
```
